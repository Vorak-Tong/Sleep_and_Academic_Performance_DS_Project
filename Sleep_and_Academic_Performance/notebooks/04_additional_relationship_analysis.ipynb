# Task 4: Additional Relationship Analysis
In this notebook, we are going to explore the relationships between:
- Social Hours and GPA
- Stress Levels and GPA


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set_theme(style="whitegrid")

# Load the cleaned dataset
df = pd.read_csv("../data/processed/student_lifestyle_cleaned.csv")
df.head()
```

### 1. Social Hours vs GPA
We use a scatter plot to analyze the relationship between social hours per day and academic performance (GPA).

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Social_Hours_Per_Day', y='GPA', alpha=0.6, color='royalblue')

plt.title('Relationship Between Social Hours Per Day and GPA', fontsize=14)
plt.xlabel('Social Hours Per Day', fontsize=12)
plt.ylabel('GPA', fontsize=12)
plt.show()
```

**Analysis:**
- Observe the trend of the data points. Is it positive, negative, or random? 
- A general hypothesis might be that extreme amounts of social hours are negatively correlated with GPA, whereas a moderate amount does not significantly hinder academic performance.

### 2. Stress Level vs GPA
Since `Stress_Level` in this dataset is actually a categorical variable (Low, Moderate, High) rather than continuous continuous hours, we will use a box plot to examine the distribution of GPA across different stress categories.

```python
plt.figure(figsize=(10, 6))
stress_order = ['Low', 'Moderate', 'High']
sns.boxplot(data=df, x='Stress_Level', y='GPA', order=stress_order, palette='Set2')

plt.title('GPA Distribution Across Different Stress Levels', fontsize=14)
plt.xlabel('Stress Level', fontsize=12)
plt.ylabel('GPA', fontsize=12)
plt.show()
```

**Analysis:**
- By comparing the medians (the horizontal lines inside the boxes), we can determine which stress level is associated with higher GPAs typically.
- Look at the spread of the boxes (interquartile range) and the "whiskers" extending out: these indicate how varied the GPA is within that stress category.
- It's possible that a "Moderate" stress level acts as a positive motivator (eustress), while a "High" stress level might be detrimental.
