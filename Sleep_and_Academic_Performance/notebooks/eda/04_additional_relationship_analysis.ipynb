# Task 4: Additional Relationship Analysis
In this notebook, we are going to explore the relationships between:
- Social Hours and GPA
- Stress Levels and GPA


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set_theme(style="whitegrid")

# Load the cleaned dataset
df = pd.read_csv("../../data/processed/student_lifestyle_cleaned.csv")
df.head()
```

### 1. Social Hours vs GPA
We use a scatter plot to analyze the relationship between social hours per day and academic performance (GPA).

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Social_Hours_Per_Day', y='GPA', alpha=0.6, color='royalblue')

plt.title('Relationship Between Social Hours Per Day and GPA', fontsize=14)
plt.xlabel('Social Hours Per Day', fontsize=12)
plt.ylabel('GPA', fontsize=12)
plt.show()
```

**Analysis:** <br>
The scatter plot shows no clear relationship between social hours per day and GPA. The data points are widely scattered without a visible upward or downward trend, indicating a very weak or negligible correlation. This suggests that the amount of time students spend on social activities does not significantly influence their academic performance in this dataset.

### 2. Stress Level vs GPA
Since `Stress_Level` in this dataset is actually a categorical variable (Low, Moderate, High) rather than continuous continuous hours, we will use a box plot to examine the distribution of GPA across different stress categories.

```python
plt.figure(figsize=(10, 6))
stress_order = ['Low', 'Moderate', 'High']
sns.boxplot(data=df, x='Stress_Level', y='GPA', order=stress_order, palette='Set2')

plt.title('GPA Distribution Across Different Stress Levels', fontsize=14)
plt.xlabel('Stress Level', fontsize=12)
plt.ylabel('GPA', fontsize=12)
plt.show()
```

**Analysis:** <br>
The boxplot shows that GPA tends to increase with higher stress levels. Students in the high stress group have the highest median GPA, followed by moderate stress, while low stress students have the lowest median GPA. This suggests a positive relationship between stress level and academic performance in this dataset. One possible explanation is that a certain level of stress may motivate students to study more and perform better. However, this does not imply causation, and further analysis is needed to understand the underlying factors.
