# Task 2: Distribution Analysis

**1. Load Cleaned Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../../data/processed/student_lifestyle_cleaned.csv")
df.head()
```

**2. Distribution of Sleep Hours**

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Sleep_Hours_Per_Day"], kde=True, bins=20, color="steelblue")
plt.title("Distribution of Sleep Hours Per Day")
plt.xlabel("Sleep Hours Per Day")
plt.ylabel("Frequency")
plt.show()

print("Key Findings - Sleep Hours Per Day:")
print(f"  Mean:     {df['Sleep_Hours_Per_Day'].mean():.2f} hours")
print(f"  Median:   {df['Sleep_Hours_Per_Day'].median():.2f} hours")
print(f"  Std Dev:  {df['Sleep_Hours_Per_Day'].std():.2f} hours")
print(f"  Min:      {df['Sleep_Hours_Per_Day'].min():.2f} hours")
print(f"  Max:      {df['Sleep_Hours_Per_Day'].max():.2f} hours")
print(f"  Skewness: {df['Sleep_Hours_Per_Day'].skew():.2f}")
```

**Explanation:** <br>
The distribution of sleep hours is relatively evenly spread between 5 and 10 hours per day, with an average of approximately 7.5 hours. The near-zero skewness indicates that the data is fairly symmetric, meaning there is no strong tendency toward very low or very high sleep durations. This suggests that students in the dataset have diverse but balanced sleep habits. Since most students sleep around the recommended 7–8 hours, this provides a good basis to further analyze whether variations in sleep duration have a measurable impact on academic performance (GPA).

**3. Distribution of GPA**

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["GPA"], kde=True, bins=20, color="coral")
plt.title("Distribution of GPA")
plt.xlabel("GPA")
plt.ylabel("Frequency")
plt.show()

print("Key Findings - GPA:")
print(f"  Mean:     {df['GPA'].mean():.2f}")
print(f"  Median:   {df['GPA'].median():.2f}")
print(f"  Std Dev:  {df['GPA'].std():.2f}")
print(f"  Min:      {df['GPA'].min():.2f}")
print(f"  Max:      {df['GPA'].max():.2f}")
print(f"  Skewness: {df['GPA'].skew():.2f}")
```

**Explanation:** <br>
The GPA distribution follows an approximately normal (bell-shaped) pattern, centered around a mean of 3.12. The mean and median are very close, and the skewness is near zero, indicating that the data is symmetric. Most students have GPAs between 2.9 and 3.3, with fewer students at the extremes. This suggests that academic performance in the dataset is relatively balanced without significant outliers. This distribution is suitable for further analysis, as it allows us to examine how variables such as sleep duration and study habits influence variations in GPA.

**4. Distribution of Study Hours**

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Study_Hours_Per_Day"], kde=True, bins=20, color="mediumseagreen")
plt.title("Distribution of Study Hours Per Day")
plt.xlabel("Study Hours Per Day")
plt.ylabel("Frequency")
plt.show()

print("Key Findings - Study Hours Per Day:")
print(f"  Mean:     {df['Study_Hours_Per_Day'].mean():.2f} hours")
print(f"  Median:   {df['Study_Hours_Per_Day'].median():.2f} hours")
print(f"  Std Dev:  {df['Study_Hours_Per_Day'].std():.2f} hours")
print(f"  Min:      {df['Study_Hours_Per_Day'].min():.2f} hours")
print(f"  Max:      {df['Study_Hours_Per_Day'].max():.2f} hours")
print(f"  Skewness: {df['Study_Hours_Per_Day'].skew():.2f}")
```

**Explanation:** <br>
The distribution of study hours is relatively evenly spread between 5 and 10 hours per day, with an average of approximately 7.5 hours. The mean and median are very close, and the skewness is near zero, indicating a symmetric distribution. This suggests that students in the dataset have varied but balanced study habits without a strong concentration in a specific range. Since study hours are directly related to academic effort, this variable is expected to have a strong influence on GPA and will be an important factor in further analysis.
